```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data
total_losses = np.load('basic_ac_total_losses.npy')
total_rewards = np.load('basic_ac_total_rewards.npy')

# Create a figure and axis
plt.figure(figsize=(12, 6))

# Plot total rewards
plt.subplot(1, 2, 1)
plt.plot(total_rewards, label='Total Rewards', color='blue')
plt.title('Total Rewards Over Episodes')
plt.xlabel('Episodes')
plt.ylabel('Total Rewards')
plt.legend()

# Add trend line for total rewards
z_rewards = np.polyfit(range(len(total_rewards)), total_rewards, 3)
p_rewards = np.poly1d(z_rewards)
plt.plot(range(len(total_rewards)), p_rewards(range(len(total_rewards))), color='black', linestyle='--', label='Trend Line')
plt.legend()

# Plot total losses
plt.subplot(1, 2, 2)
plt.plot(total_losses, label='Total Losses', color='red')
plt.title('Total Losses Over Episodes')
plt.xlabel('Episodes')
plt.ylabel('Total Losses')
plt.legend()

# Add trend line for total losses
z_losses = np.polyfit(range(len(total_losses)), total_losses, 3)
p_losses = np.poly1d(z_losses)
plt.plot(range(len(total_losses)), p_losses(range(len(total_losses))), color='black', linestyle='--', label='Trend Line')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
