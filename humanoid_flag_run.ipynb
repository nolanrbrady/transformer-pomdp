```python
import gymnasium as gym
# Import the VizDoom gymnasium wrapper so that the VizDoom environments are registered
from vizdoom import gymnasium_wrapper  
import numpy as np
from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env
```

```python
# Define the environment id.
# Note: Depending on your build/installation, VizDoom environments are often registered with a -v0 suffix.
env_id = "VizdoomCorridor-v0"

# Create a vectorized environment using make_vec_env.
# Here we create 4 parallel environments.
env = make_vec_env(env_id, n_envs=4)

# Since VizDoom produces image observations, we use a CNN policy.
model = PPO("MultiInputPolicy", env, verbose=1)

# Train the model; adjust total_timesteps as needed.
model.learn(total_timesteps=100000)

# Save the trained model.
model.save("ppo_vizdoom_corridor")
print("Model saved as ppo_vizdoom_corridor")

env.close()

# Run a demonstration rollout with a video
env = gym.make(env_id, render_mode="human")
obs, info = env.reset()

# Inference loop: the environment will automatically render to a window.
while True:
    action, _ = model.predict(obs, deterministic=True)  # Replace with your trained model
    obs, reward, terminated, truncated, info = env.step(action)
    
    # Calling render() is optional in human mode as the window updates automatically.
    env.render()
    
    # If the episode ends, reset the environment.
    if terminated or truncated:
        obs, info = env.reset()

env.close()
```
